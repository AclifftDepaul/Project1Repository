```python
import os

# Module for reading CSV files
import csv
import pandas as pd
import matplotlib.pyplot as plt



file_one = "../Resources/data-table.csv"
file_two = "../Resources/Obesity-prevalence-by-state-2021.csv"

ObesityFullDF = pd.read_csv(file_two, delimiter = ",")


CancerAreaDF = pd.read_csv(file_one, delimiter = ",")
ObesityFullDF = ObesityFullDF.dropna()
CancerAreaDF = CancerAreaDF.dropna()
ObesityFullDF
```

```python
YearCancerAreaDF = CancerAreaDF.loc[(CancerAreaDF["YEAR"] == 2021)]
YearCancerAreaDF

YearCancerAreaDF['DEATHS'] = YearCancerAreaDF['DEATHS'].astype(int)
```

```python
state_mapping = {
    'AL': 'Alabama',
    'AK': 'Alaska',
    'AZ': 'Arizona',
    'AR': 'Arkansas',
    'CA': 'California',
    'CO': 'Colorado',
    'CT': 'Connecticut',
    'DE': 'Delaware',
    'FL': 'Florida',
    'GA': 'Georgia',
    'HI': 'Hawaii',
    'ID': 'Idaho',
    'IL': 'Illinois',
    'IN': 'Indiana',
    'IA': 'Iowa',
    'KS': 'Kansas',
    'KY': 'Kentucky',
    'LA': 'Louisiana',
    'ME': 'Maine',
    'MD': 'Maryland',
    'MA': 'Massachusetts',
    'MI': 'Michigan',
    'MN': 'Minnesota',
    'MS': 'Mississippi',
    'MO': 'Missouri',
    'MT': 'Montana',
    'NE': 'Nebraska',
    'NV': 'Nevada',
    'NH': 'New Hampshire',
    'NJ': 'New Jersey',
    'NM': 'New Mexico',
    'NY': 'New York',
    'NC': 'North Carolina',
    'ND': 'North Dakota',
    'OH': 'Ohio',
    'OK': 'Oklahoma',
    'OR': 'Oregon',
    'PA': 'Pennsylvania',
    'RI': 'Rhode Island',
    'SC': 'South Carolina',
    'SD': 'South Dakota',
    'TN': 'Tennessee',
    'TX': 'Texas',
    'UT': 'Utah',
    'VT': 'Vermont',
    'VA': 'Virginia',
    'WA': 'Washington',
    'WV': 'West Virginia',
    'WI': 'Wisconsin',
    'WY': 'Wyoming'
}

YearCancerAreaDF['STATE'] = YearCancerAreaDF['STATE'].map(state_mapping)
```

```python
obesity_cancer_df = pd.merge(YearCancerAreaDF, ObesityFullDF, left_on='STATE', right_on='State', how='inner', suffixes=('', '_drop'))
obesity_cancer_df = obesity_cancer_df.drop(columns=['State'])
obesity_cancer_df = obesity_cancer_df[obesity_cancer_df['Prevalence'] != 'No data']

# Create the scatter plot


obesity_cancer_df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have the 'ObesityFullDF' DataFrame already loaded

# Create a bar graph for obesity prevalence by state
ObesityFullDF = ObesityFullDF.sort_values(by='Prevalence', ascending=True)
ObesityFullDF_filtered = ObesityFullDF.query("State != 'Florida'")

plt.figure(figsize=(12, 10))
plt.bar(ObesityFullDF['State'], ObesityFullDF['Prevalence'], color='skyblue')
plt.title('Obesity Prevalence by State (2021)')
plt.xlabel('State')
plt.ylabel('Prevalence (%)')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Sort the DataFrame by RATE (cancer rate) in ascending order for a clearer bar chart
obesity_cancer_df = obesity_cancer_df.sort_values(by='RATE', ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.barh(obesity_cancer_df['STATE'], obesity_cancer_df['RATE'], color='skyblue')
plt.title('Cancer Rate by State in 2021')
plt.xlabel('Cancer Rate')
plt.ylabel('State')
plt.gca().invert_yaxis()  # Reverse the order to have the highest rates at the top
plt.grid(axis='x')  # Add a grid for the x-axis

# Annotate the bar chart with the RATE values
for i, rate in enumerate(obesity_cancer_df['RATE']):
    plt.text(rate, i, f' {rate:.1f}', va='center', fontsize=10)

plt.show()
```

```python

rate = obesity_cancer_df['RATE'].astype(float)
prevalence = obesity_cancer_df['Prevalence'].astype(float)

plt.figure(figsize=(10, 6))
plt.scatter(prevalence, rate, c='b', alpha=0.5)
plt.title('Cancer Rate vs. Obesity Prevalence')
plt.xlabel('Prevalence (%)')
plt.ylabel('Cancer Rate')
plt.grid(True)

# Annotate the data points with state names
for i, state in enumerate(obesity_cancer_df['STATE']):
    plt.annotate(state, (prevalence.iloc[i], rate.iloc[i]), fontsize=8, alpha=0.7)

plt.show()
```
