```python
import os

# Module for reading CSV files
import csv
import pandas as pd


file_one = "../Resources/Adult_Tobacco_Consumption_In_The_U.S.__2000-Present_20231019.csv"
file_two = "../Resources/data-table.csv"


DrugFullDF = pd.read_csv(file_one, delimiter = ',' )
CancerRate = pd.read_csv(file_two, delimiter = ',' )

DrugFullDF = DrugFullDF.dropna()
CancerRate = CancerRate.dropna()

CancerRate['deaths'] = [str(x).strip(',')[-3] for x in CancerRate.iloc[:,3]]

CancerRate["deaths"] = CancerRate["deaths"].astype('int64')

CancerRate
```

```python
DrugNonCombustibleDF = DrugFullDF.loc[DrugFullDF['Topic'] == 'Noncombustible Tobacco']
DrugNonCombustibleDF.head()
```

```python
TotalDrugNonCombustibleDF = DrugNonCombustibleDF.groupby(['Year']).sum()
TotalDrugNonCombustibleDF.head()
```

```python
DrugFullDFYear = DrugFullDF.groupby(['Year']).sum()
DrugFullDFYear = DrugFullDFYear.drop("Population", axis='columns')
DrugFullDFYear
```

```python
CancerRate = CancerRate.groupby(['YEAR']).sum()
CancerRate
```

```python
import matplotlib.pyplot as plt

#creating a line plot
plt.figure(figsize=(10, 6))
plt.plot(CancerRate.index, CancerRate['deaths'], marker='o', linestyle='-', color='b')
plt.title('Cancer Deaths Over Time')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.grid(True)

# Displaying the line plot
plt.show()
```

```python
#Creating a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(DrugFullDFYear.index, DrugFullDFYear['Total'], c='b', marker='o', s=50)
plt.title('Scatter Plot of Drug Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Total Consumption')
plt.grid(True)

# Displaying the scatter plot
plt.show()
```
