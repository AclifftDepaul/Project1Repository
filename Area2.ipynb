```python
import os

# Module for reading CSV files
import csv
import pandas as pd


file_one = "Resources/data-table.csv"

USCSAreaDF = pd.read_csv(file_one, delimiter = ",")

USCSAreaDF = USCSAreaDF.dropna()
USCSAreaDF
```

```python
USCSAreaDF["DEATHS"] = USCSAreaDF["DEATHS"].str.replace(',','').astype(int)
print(USCSAreaDF.dtypes)
```

```python
GroupedUSCSAreaDF = USCSAreaDF.groupby(['STATE'])

SumUSCSAreaDF = GroupedUSCSAreaDF[['DEATHS']].mean().sum()
SumUSCSAreaDF
```
